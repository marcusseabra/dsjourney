```python
dados_tabela = pagina.extract_table()
for linhas in dados_tabela[1:]:
    # print(linhas)
    texto = ""
    
    nomeAtributo = linhas[1]
    nomeAtributo = nomeAtributo.replace(" de ", " ")
    nomeAtributo = nomeAtributo.replace(" do ", " ")
    nomeAtributo = nomeAtributo.replace(" da ", " ")
    nomeAtributo = nomeAtributo.replace(" dos ", " ")
    nomeAtributo = nomeAtributo.replace(" das ", " ")
    nomeAtributo = nomeAtributo.replace(" no ", " ")
    nomeAtributo = nomeAtributo.replace(" na ", " ")
    nomeAtributo = nomeAtributo.replace(" nos ", " ")
    nomeAtributo = nomeAtributo.replace(" nas ", " ")
    nomeAtributo = nomeAtributo.replace("\nde ", " ")
    nomeAtributo = nomeAtributo.replace("\ndo ", " ")
    nomeAtributo = nomeAtributo.replace("\nda ", " ")
    nomeAtributo = nomeAtributo.replace("\ndos ", " ")
    nomeAtributo = nomeAtributo.replace("\ndas ", " ")
    nomeAtributo = nomeAtributo.replace("\nno ", " ")
    nomeAtributo = nomeAtributo.replace("\nna ", " ")
    nomeAtributo = nomeAtributo.replace("\nnos ", " ")
    nomeAtributo = nomeAtributo.replace("\nnas ", " ")
    nomeAtributo = nomeAtributo.replace("(os)", "")
    nomeAtributo = nomeAtributo.replace("(as)", "")
    nomeAtributo = nomeAtributo.replace("(es)", "")
    nomeAtributo = nomeAtributo.replace("(s)", "")
    nomeAtributo = nomeAtributo.replace("ã", "a")
    nomeAtributo = nomeAtributo.replace("á", "a")
    nomeAtributo = nomeAtributo.replace("à", "a")
    nomeAtributo = nomeAtributo.replace("â", "a")
    nomeAtributo = nomeAtributo.replace("é", "e")
    nomeAtributo = nomeAtributo.replace("ê", "e")
    nomeAtributo = nomeAtributo.replace("í", "i")
    nomeAtributo = nomeAtributo.replace("ô", "o")
    nomeAtributo = nomeAtributo.replace("ó", "o")
    nomeAtributo = nomeAtributo.replace("ú", "u")
    nomeAtributo = nomeAtributo.replace("ü", "u")
    nomeAtributo = nomeAtributo.replace("ç", "c")
    nomeAtributo = nomeAtributo.replace("\n", " ")
    nomeAtributo = nomeAtributo.replace("–", "")
    nomeAtributo = nomeAtributo.replace("/", "")

    # print(nomeAtributo)

    ordem = int(linhas[0])
    tamanho = int(linhas[3])
    
    if nomeAtributo.find("Data") >= 0:
        texto = "private Date"
    elif nomeAtributo.find("Valor") >= 0 or nomeAtributo.find("Saldo") >= 0:
        texto = "private Double"
    else:
        if linhas[4] == "Numérico":
            texto = "private Integer"
        if linhas[4] == "Alfanumérico" or linhas[4] == "Alfabético":
            texto = "private String"
    
    nomeAtributo = nomeAtributo.replace(" ", "")
    nomeAtributo = nomeAtributo[0].lower() + nomeAtributo[1:]
    
    print("@CampoParaTexto(ordem = " + str(ordem) 
          + ", tamanho = " + str(tamanho) 
          + ")\n" + texto + " " + nomeAtributo + ";\n")
```
